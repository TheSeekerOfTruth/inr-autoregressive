```python
import torch
import pandas as pd

# 1. Define the path to your file
model_path = 'data/mnist_png_training_0_1.pt'  # or 'your_model.pt'

# 2. Load the weights safely
# 'weights_only=True' is recommended for security unless you are loading custom classes
try:
    checkpoint = torch.load(model_path, map_location=torch.device('cpu'), weights_only=True)
except Exception as e:
    print("Trying without weights_only due to an error...")
    checkpoint = torch.load(model_path, map_location=torch.device('cpu'))

# 3. Check what type of object was saved
# Sometimes it's the raw state_dict, sometimes it's a dict containing 'state_dict', 'epoch', etc.
if isinstance(checkpoint, dict):
    if 'state_dict' in checkpoint:
        state_dict = checkpoint['state_dict']
        print("Found 'state_dict' key inside the checkpoint dictionary.\n")
    else:
        state_dict = checkpoint
        print("Loaded checkpoint directly as state_dict.\n")
else:
    # If they saved the whole model object instead of just the weights
    state_dict = checkpoint.state_dict()
    print("Loaded a full model object; extracted state_dict.\n")

# 4. Explicitly view how the weights look
print(f"{'Layer Name':<50} | {'Shape':<20} | {'Data Type'}")
print("-" * 90)

for layer_name, weights in state_dict.items():
    print(f"{layer_name:<50} | {str(list(weights.shape)):<20} | {weights.dtype}")
    
    # Print the actual raw weight tensor values
    print("Raw Weights/Values:")
    print(weights)
    print("=" * 90 + "\n")
```

```python
# Create a summary table of the layers
layer_data = []
for name, tensor in state_dict.items():
    layer_data.append({
        "Layer Name": name,
        "Shape": list(tensor.shape),
        "Data Type": tensor.dtype,
        "Sample Values (First 3)": tensor.flatten()[:33].tolist() # flattens and grabs first 3 values
    })

df = pd.DataFrame(layer_data)
df
```
